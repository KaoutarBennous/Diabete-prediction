```python
import pandas as pd
```

```python
# read the csv-formatted data file into a pandas dataframe
df=pd.read_csv('diabetes.csv')
# get shape of data frame
print('Shape (n_rows,n_columns) of dataframe:',df.shape)
# print top 5 rows (default) of the data frame
df.head()
```

```python
import seaborn as sns
sns.countplot(df['Outcome'],label="Count")
```

```python
import matplotlib.pyplot as plt
# make it interactive in the notebook (Only needed for that librairy)
%matplotlib inline 
plt.figure()
plt.hist(df[df.Outcome==1]['Age'],bins=15,label='Diabetes',color='r',alpha=0.2)
plt.hist(df[df.Outcome==0]['Age'],bins=15,label='No Diabetes',color='b',alpha=0.2)
plt.xlabel('Age')
plt.ylabel('Frequence')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.hist(df[df.Outcome==1]['Glucose'],bins=15,label='Diabetes',color='r',alpha=0.2)
plt.hist(df[df.Outcome==0]['Glucose'],bins=15,label='No Diabetes',color='b',alpha=0.2)
plt.xlabel('Glucose')
plt.ylabel('Frequence')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.hist(df[df.Outcome==1]['DiabetesPedigreeFunction'],bins=15,label='Diabetes',color='r',alpha=0.2)
plt.hist(df[df.Outcome==0]['DiabetesPedigreeFunction'],bins=15,label='No Diabetes',color='b',alpha=0.2)
plt.xlabel('DiabetesPedigreeFunction')
plt.ylabel('Frequence')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.hist(df[df.Outcome==1]['BloodPressure'],bins=15,label='Diabetes',color='r',alpha=0.2)
plt.hist(df[df.Outcome==0]['BloodPressure'],bins=15,label='No Diabetes',color='b',alpha=0.2)
plt.xlabel('BloodPressure')
plt.ylabel('Frequence')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.hist(df[df.Outcome==1]['BMI'],bins=15,label='Diabetes',color='r',alpha=0.2)
plt.hist(df[df.Outcome==0]['BMI'],bins=15,label='No Diabetes',color='b',alpha=0.2)
plt.xlabel('BMI')
plt.ylabel('Frequence')
plt.legend()
plt.show()
```

```python
corr_mat=df.corr()
sns.heatmap(corr_mat, annot=True)
```

```python
# plot Glucose vs BloodPressure and color points according to Outcome
plt.figure()
plt.scatter(df[df.Outcome==1].Glucose,df[df.Outcome==1].BloodPressure,label='Diabetes',color='r',s=3)
plt.scatter(df[df.Outcome==0].Glucose,df[df.Outcome==0].BloodPressure,label='No Diabetes',color='b',s=3)
plt.legend()
plt.xlabel('Glucose')
plt.ylabel('BloodPressure')
```
